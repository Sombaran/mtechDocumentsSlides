**For Loop Examples**

```python
1. #Basic For Loop#
# for (int i=0;i<5;i++)
#[2:5]
for i in range(12):
   print(i)
```

```python
2. #For Loop with List#

fruits = ["apple", "banana", "cherry"]
for i in fruits:
    print(i)
```

```python
2. #For Loop with Tuple#

fruits = ("apple", "banana", "cherry")
for i in fruits:
    print(i)
```

```python
3. #For Loop with Range and Step#

for i in range(10,25,3):
 print(i)
```

```python
3. #For Loop with Range and Step#

for i in range(25,2,-11):
 print(i)
```

```python
number = int(input("Enter a number"))

for i in range(number,number*12+1,number):
  print(i)
```

```python
number = int(input("Enter a number: "))

for i in range(13):
  print(f"{number} * {i} = {number*i}")
```

```python
4. #For Loop with String#

greeting = ["hello"]
for char in greeting:
 print("Read: ", char)

for char in "hello":
 print("Character Read: ", char)

print("*#"*25)
for char in "h":
 print("Character Read: ", char)
```

```python
5. #For Loop with Dictionary#
#key-value
ages = {"Alice": 25, "Bob": 30, "Charlie": 25}
for name, age in ages.items():
 print(name, age)
```

```python

my_dict = {"a": 10, "b": 22, "c": -3, "d":710}

for key in my_dict.keys():
  print(key)
```

```python
#dictionary = key-value
my_dict = {"a": 10, "b": 22, "c": -3, "d":710}

for key in my_dict:
    print(key)
print("*"*30)
for key in my_dict.keys():
    print(key)
print("*"*30)
for value in my_dict.values():
    print(value)
print("*"*30)
for x, y in my_dict.items():
 print(x, y)
print("*"*30)


```

```python
6. #For Loop with Tuple#
a,b = 4,5
print(a,b)

coordinates = [(1, 2), (3, 4), (5, 6)]
for x, y in coordinates:
 print(f"x: {x}, y: {y}")
```

```python
7. #For Loop with List Comprehension#

numbers = [1, 2, 3, 4, 5]

squares = [x**2 for x in numbers]
print(squares)

sqaures = []
for x in numbers:
  sqaures.append(x**2)
print(sqaures)
```

```python
8. #For Loop with Enumerate#

items = ["a", "b", "c"]
index = 0
for item in items:
 print(f"At Index: {index}, the value is {item} ")
 index += 1
print("*************")

for index, item in enumerate(items):
 print(index, item)
```

```python
9. #For Loop with Nested Loops#
count = 0
for i in range(3): #m times
  for j in range(5): # n times
     print(f"i: {i}, j: {j}")  # m * n times
     count += 1

  print("***************")
print(count)
```

```python
10. #For Loop with Conditional Statement#

numbers = [1, 2, 3, 4, 5]
for num in numbers:
  if num % 2 == 0:
    print(num, "is even")
  else:
    print(num, "is odd")
```

**While Loop Examples**

```python
1. #Basic While Loop#

for i in range(0,5,1):
  print(i)

print("*"*23)

i = 0
while i < 5:
 print(i)
 i += 1
```

```python
2. #While Loop with List#

numbers = [1, 2, 3]
index = 0
while index < len(numbers):
 print(numbers[index])
 index += 1
```

```python
3. #While Loop with Break#

count = 0
while True:
 print(count)
 count += 1
 if count == 13:
  break

print("End of Program")
```

```python
4. #While Loop with Continue#

count = 0
while count < 5:
 count += 1
 if count % 2 == 0:
  continue
 print(count)
```

```python
5. #While Loop with Conditional Check#

number = 5
while number > 0:
 print(number)
 number -= 1
```

```python
6. #While Loop with User Input#

user_input = ""
while user_input.lower() != "quit":
 user_input = input("Type 'quit' to stop: ")
 print("You typed:", user_input)
```

```python
7. #While Loop with List#

numbers = [10, 20, 30]
index = 0
while index < len(numbers):
 print(numbers[index])
 index += 1
```

```python
8. #While Loop with Infinite Loop and Break#

while True:
 user_input = input("Enter something (or 'stop' to quit): ")
 if user_input == "stop":
  break
 print("You entered:", user_input)
```

```python
9. #While Loop with Nested Loops#

i = 0
while i < 3:
  j = 0
  while j < 2:
   print(f"i: {i}, j: {j}")
   j += 1
  i += 1
```

```python
10. #While Loop with Accumulation#

total = 0
number = 1
count=0
while number <= 5:
  total = total + number
  number += 1
print("Sum:", total)
```
