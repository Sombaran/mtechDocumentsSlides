```python
1. #Basic AND Operator#

a = True
b = False
print(type(a))
result = a and b  # False
print(result)
```

```python
2. #Basic OR Operator#

a = True
b = False
result = a or b  # True
print(result)
```

```python
3. #Basic NOT Operator#

a = True
result = not a  # False
print(result)
```

```python
4. #Combining AND and OR#

a = True
b = False
c = True
result = a and (b or c)  # True
# a and (b or c)
# T and (F or T)
# T and (T)
print(result)
```

```python
5. #Combining AND, OR, and NOT#

a = True
b = False
result = not (a and b)  # True
print(result)
```

```python
6. #Check if Both Conditions are True#

age = 25
is_student = True
result = age > 18 and is_student  # True
print(result)
```

```python
7. #Check if Either Condition is True#

age = 17
is_student = True
result = age > 18 or is_student  # True
print(result)
```

```python
8. #Check if Both Conditions are False#

age = 17
is_student = False
result = not (age > 18 or is_student)  # True
print(result)
```

```python
9. #Check Multiple Conditions#

temperature = 25
raining = False
result = temperature > 20 and not raining  # True

#temperature > 20 and not raining
# (25 > 20) and (not raining)
# (T) and (not False)
# (T) and (T)

print(result)
```

```python
10. #Logical Operators with Integers#

x = 10
y = 20
result = (x < 15) and (y > 15)  # True
print(result)
```

```python
11. #Check if Variable is in Range#

number = 10
result = (number > 5) and (number < 15)  # True
print(result)
```

```python
12. #Negating a Condition#

value = 10
result = not (value == 10)  # False
print(result)
```

```python
13. #Logical Operators with Strings#

name = "Alice"
age = 15
result = (name == "Alice") and (age > 20)  # True
print(result)
```

```python
14. #Logical Operators with Lists#

list1 = [1, 2, 3]
list2 = [4, 5, 6]
result = (len(list1) > 2) or (len(list2) < 3)  # True
print(result)
```

```python
15. #Using Logical Operators in a Function#

def check_conditions(a, b):
  return (a > 0) and (b < 10)

result = check_conditions(5, 8)  # True
print(result)
```

```python
16. #Logical Operators with Comparisons#

x = 5
y = 10
result = (x == 5) or (y == 5)  # True
print(result)
```

```python
17. #Combining NOT with AND#

a = True
b = False
result = not (a and b)  # True
print(result)
```

```python
18. #Combining NOT with OR#

a = True
b = False
result = not (a or b)  # False
print(result)
```

```python
19. #Using Logical Operators in Conditional Statements#

score = 85
result = (score > 90) or (score == 85)  # True
print(result)
```

```python
20. #Logical Operators in List Comprehensions#

numbers = [1, 2, 3, 4, 5]
result = [num for num in numbers if (num > 2) and (num < 5)]  # [3, 4]
print(result)
```

```python

#result = [num for num in numbers if (num > 2) and (num < 5)]  # [3, 4]

numbers = [1, 2, 3, 4, 5]
result = []

for i in numbers:
    if (i > 2) and (i < 5):
        result.append(i)

print(result)
```
