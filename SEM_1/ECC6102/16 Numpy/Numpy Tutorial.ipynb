```python
import numpy as np

# Create a 1D array from a Python list
arr = np.array([1, 2, 3, 4, 5])
print(arr, type(arr))

print("*"*30)

# Create a 2D array (matrix) of zeros
zeros = np.zeros((3, 3))
print(zeros)

print("*"*30)
# Create a 1D array with values ranging from 0 to 9
arr_range = np.arange(10)
print(arr_range)
```

```python
arr = np.array([10, 20, 30, 40, 50])

# Get a slice of the array
slice_arr = arr[1:4]  # Elements from index 1 to 3
print(slice_arr)

# Access individual elements
first_element = arr[0]
print(first_element)


# Access individual elements
last_element = arr[-1]
print(last_element)
```

```python
arr = np.arange(12)  # Create an array with values from 0 to 11
print(arr)


print("*"*30)
# Reshape it into a 3x4 matrix
reshaped_arr = arr.reshape(3, 4)
print(reshaped_arr)


print("*"*30)
# Reshape it into a 6x2 matrix
reshaped_arr = arr.reshape(6, 2)
print(reshaped_arr, type(reshaped_arr))
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("*"*30)
# Element-wise addition
sum_arr = arr1 + arr2
print(sum_arr)

print("*"*30)
# Element-wise multiplication
product_arr = arr1 * arr2
print(product_arr)

print("*"*30)
# Scalar multiplication (every element is multiplied by 2)
scaled_arr = arr1 * 2
print(scaled_arr)
print(arr1)
```

```python
arr = np.array([10, 20, 30, 40, 50])

print("*"*30)
# Sum of all elements
sum_value = np.sum(arr)
print(sum_value)

print("*"*30)
# Mean of the elements
mean_value = np.mean(arr)
print(mean_value)

print("*"*30)
# Maximum and Minimum
max_value = np.max(arr)
min_value = np.min(arr)
print(f"Max: {max_value}, Min: {min_value}")
```

```python
# Create two 2x2 matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Perform matrix multiplication using np.dot or @ operator
result = np.dot(matrix1, matrix2)  # Same as: result = matrix1 @ matrix2
print(result)
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])

# Transpose the matrix
transposed = np.transpose(matrix)
# Alternatively: transposed = matrix.T
print(transposed)
```

```python
# Generate a 2x3 array with random values between 0 and 1
random_arr = np.random.rand(2, 3)
print(random_arr)

# Generate an array with random integers between 0 and 10
random_ints = np.random.randint(0, 10, size=(9, 9))
print(random_ints)
```

```python
# Create two 1D arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Concatenate them along axis 0 (row-wise)
concatenated = np.concatenate((arr1, arr2))
print(concatenated)


print("*"*30)
# Create two 2D arrays and concatenate along axis 1 (column-wise)
arr_2d_1 = np.array([[1, 2], [3, 4]])
arr_2d_2 = np.array([[5, 6], [7, 8]])

print("*"*30)
concatenated_2d = np.concatenate((arr_2d_1, arr_2d_2), axis=0)
print(concatenated_2d)


print("*"*30)
concatenated_2d = np.concatenate((arr_2d_1, arr_2d_2), axis=1)
print(concatenated_2d)

arr_2d_1 = np.array([[1, 2, 3], [99, 4, 5]])
arr_2d_2 = np.array([[5, 6, 7], [77, 88, 95]])

print("*"*30)
concatenated_2d = np.concatenate((arr_2d_1, arr_2d_2), axis=0)
print(concatenated_2d)


print("*"*30)
concatenated_2d = np.concatenate((arr_2d_1, arr_2d_2), axis=1)
print(concatenated_2d)
```

```python
import numpy as np
import time

# Create large arrays/lists
size = 10**6
print(size)
python_list1 = list(range(size))
python_list2 = list(range(size))

numpy_array1 = np.arange(size)
numpy_array2 = np.arange(size)

# Timing addition using Python lists
start_time = time.time()
result_list = [x + y for x, y in zip(python_list1, python_list2)]
python_time = time.time() - start_time

# Timing addition using NumPy arrays
start_time = time.time()
result_numpy = numpy_array1 + numpy_array2
numpy_time = time.time() - start_time

# Print results
print(f"Time taken using Python lists: {python_time:.6f} seconds")
print(f"Time taken using NumPy arrays: {numpy_time:.6f} seconds")
```

```python
import numpy as np

arr = np.array([0, np.pi/2, np.pi])
print(len(arr))
# Apply sine function element-wise
sin_values = np.sin(arr)
print("Sine values:", sin_values)


cos_values = np.cos(arr)
print("Sine values:", cos_values)

# Apply exponential and square root
exp_values = np.exp(arr)
sqrt_values = np.sqrt(np.array([1, 4, 91]))

print("Exponential values:", exp_values)
print("Square root values:", sqrt_values)
```
