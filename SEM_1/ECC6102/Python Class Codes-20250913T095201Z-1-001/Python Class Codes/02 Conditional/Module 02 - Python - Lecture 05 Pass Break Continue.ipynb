**Break Statement Examples**

```python
#1. #Breaking Out of a For Loop#

for i in range(10): # 0, ..., 9
 print(i)
 if i == 5:
  break


print("===Main", i)


print("End of Program")
```

```python
#2. #Breaking Out of a While Loop#

count = 0
while True:
 if count == 3:
  break
 print(count)
 count += 1
```

```python
#3. #Breaking from Nested Loops#

for i in range(5):
  for j in range(5):
    if i == 1 and j == 1:
      break
    print(f"i: {i}, j: {j}")
```

```python
#4. #Breaking in a While Loop with User Input#

while True:
 user_input = input("Enter 'stop' to break: ")
 if user_input.lower() == 'stop':
  break
 print("You entered:", user_input)
```

```python
#5. #Breaking from a Loop with Condition#

numbers = [1, 2, 3, 4, 5]
for number in numbers:
 if number > 3:
  break
 print(number)
```

**Continue Statement Examples**

```python
#1. #Skipping Even Numbers in a For Loop#

for i in range(10):
 if i % 2 == 0:
  print(f"Inside if Loop. i = {i}")
  continue
 print(i)
```

```python
#2. #Skipping the Current Iteration in a While Loop#

count = 0
while count < 5:
 count += 1
 if count == 3:
  continue
 print(count)
```

```python
#3. #Skipping the Rest of the Code in a Loop#

for i in range(10):
 if i == 4:
  continue
 print(i)
```

```python
#4. #Using Continue in a Nested Loop#
for i in range(3):
 for j in range(3):
  if i == j:
    break
  print(f"i: {i}, j: {j}")

print("----------------")
for i in range(3):
 for j in range(3):
  if i == j:
    continue
  print(f"i: {i}, j: {j}")
```

```python
#5. #Skipping Specific Elements in a List#

fruits = ["apple", "banana", "cherry", "date"]
for fruit in fruits:
 if fruit == "banana":
  continue
 print(fruit)
```
