```python
### 1. Basic Variable Assignment

x = 10
print(x)
print(type(x))

y = 5.5265
print(y)
print(type(y))

name = "Rahul Dravid"
print(name)
print(type(name))

is_student = True

print(is_student, type(is_student))
```

```python
### 2. Changing Variable Values

x = 20
print(x)
y = y + 2.5
print(y)
name = "Bob"
is_student = False
```

```python
### 3. Multiple Assignments

a, b, c, d  = -9, 2.5, "Hello", "India"
print(a)
print(b)
print(c)
print(d)
```

```python
### 4. String Concatenation

first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)
```

```python
### 5. Integer and Float Arithmetic

x = 5
y = 2
sum_result = x + y
div_result = x / y
print(sum_result)
print(div_result)
```

```python
### 6. Type Conversion

num_str = "700"
print(num_str, type(num_str))
num_int = int(num_str)
float_num = float(num_int)

print(num_int, type(num_int))
print(float_num, type(float_num))

# num_str = "400.5"
# num_int = int(num_str)


num_str = "400.5"
num_int = float(num_str)
print(num_int, type(num_int))
```

```python
### 7. Boolean Operations

x = 10
y = 10
is_greater = x > y  #False
print(is_greater)

is_equal = x == y
print(is_equal)

print(y>x)
```
