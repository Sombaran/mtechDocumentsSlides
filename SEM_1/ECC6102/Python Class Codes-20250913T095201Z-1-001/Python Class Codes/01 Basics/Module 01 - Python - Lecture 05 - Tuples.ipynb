```python
### tuples in Python
### tuples are Read Only. Cant be modified!

### 1. Creating a Simple Tuple

my_list  = [111, 22, 99]
my_tuple = (111, 22, 99)

print(my_list, type(my_list))
print(my_tuple, type(my_tuple))
```

```python
### 2. Accessing Tuple Elements

my_tuple = (111, 22, 99)
first_element = my_tuple[0]
second_element = my_tuple[1]
last_element = my_tuple[-1]

print(first_element)
print(second_element)
print(last_element)
```

```python
### 3. Slicing a Tuple
my_tuple = (111, 22, 99)
sub_tuple = my_tuple[1:3]
print(sub_tuple)
```

```python
my_list  = [111, 22, 99]
my_tuple = (111, 22, 99,88,34,51,9)  # Re-initialize
print(my_list)
my_list[0] = -1
print(my_list)

print(my_tuple)
my_tuple[2] = -1  #Not allowed

my_tuple = (-1, 22, 99,88,34,51,9)  # Re-initialize
print(my_tuple)
```

```python
### 4. Nested Tuples

nested_tuple = (
    (1, 2),
     (3, 4)
     )
x = nested_tuple[0]
print(x)
print(nested_tuple[1][0])
print(nested_tuple[1][1])
```

```python
### 5. Tuple Assignment

a, b, c = (1, 2, 3)
print(a,b,c)
```

```python
### 7. Concatenating Tuples

tuple1 = (1, 2)
tuple2 = (3, 4)

print(id(tuple1))
print(id(tuple2))
concatenate_tuple = tuple1 + tuple2
print(tuple1)
print(tuple2)

print(id(concatenate_tuple))
print(concatenate_tuple)

tuple1  = tuple1 + tuple2

print(id(tuple1))
print(tuple1)
```

```python
### 8. Repeating Tuple Elements

repeated_tuple = (1, 2) * 5
print(repeated_tuple)
```

```python
### 9. Tuple Unpacking

a, b, c = (1, 2, 3)
print(a,b,c)
```

```python
### 10. Counting Elements in a Tuple
my_tuple = (1, 2, 2, 3, 5, 7 ,45, 4, 2, 2)
count_of_2 = my_tuple.count(322)
print(count_of_2)
```

```python
### 11. Finding Index of an Element

index_of_45 = my_tuple.index(45)
print(index_of_45)

for x in my_tuple:
    print(x)
```

```python
### 12. Converting a List to a Tuple

my_list = [1, 2, 3]
my_tuple = tuple(my_list)

print(type(my_list), type(my_tuple))
print((my_list), (my_tuple))
```

```python
### 12. Converting a Tuple to a List

my_tuple = (1, 2, 3)
my_list = list(my_tuple)

print(type(my_list), type(my_tuple))
print((my_list), (my_tuple))
```

```python
my_tuple = (1, 2, 3)

print((my_tuple), type(my_tuple))
my_list = list(my_tuple)
print((my_list), type(my_list))

my_list[0] = 999
my_list[-1] = 'Python'

print((my_list), type(my_list))

my_tuple =  tuple(my_list)

print((my_tuple), type(my_tuple))
```

```python
### 13. Tuple Comprehension (Creating a Tuple from a Generator Expression)

squares = tuple(x**2 for x in range(5))

# for x in range(5): # 0,1,2,3,4

print(squares)


squares = list(x**2 for x in range(5))

# for x in range(5): # 0,1,2,3,4

print(squares)
```

```python
### 15. Swapping Values Using Tuples

x, y = 5, 10
print(f"Before swapping x={x} ; y={y}")
x, y = y, x
print(f"After swapping x={x} ; y={y}")
```

```python
### 16. Nested Tuples Unpacking

nested_tuple = ((1, 2), (3, 4))
((a, b), (c, d)) = nested_tuple
((a, b), (c, d)) =  ((1, 2), (3, 4))
print(a,b,c,d)
```

```python
### 17. Tuple with Mixed Data Types

mixed_tuple = (1, "hello", 3.14, True)
print(mixed_tuple)
```

```python
#18 Time betwen tuple and list.
import timeit

# Define a test case for a list and a tuple
list_test = timeit.timeit(stmt="[1, 2, 3, 4, 5]", number=10000000)
tuple_test = timeit.timeit(stmt="(1, 2, 3, 4, 5)", number=10000000)

print(f"List time: {list_test}")
print(f"Tuple time: {tuple_test}")
improvement = abs(100*(tuple_test-list_test)/tuple_test)
print(f"% Improvement: {improvement}")
```
