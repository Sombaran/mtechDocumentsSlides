```python
### 1. Creating a Simple List

fruits = ["apple", "banana", "cherry" , 3.14, 99]
numbers = [1, 2, 3, 4, 5 , "Python" , 5.234]
marks = [4, 7, 2, 8, 5]

print(fruits)
print(type(fruits))

print(numbers)
print(type(numbers))
```

```python
### 2. Accessing List Elements


print(fruits)
print(numbers)

first_fruit = fruits[0]
last_number = numbers[-1]

print(first_fruit)
print(last_number)
print(fruits[1])
```

```python
### 3. Modifying List Elements
fruits = ["apple", "banana", "cherry" , 3.14]
print(fruits)
fruits[1] = "blueberry"
print(fruits)

numbers = [1, 2, 3, 4, 5 , "Python" , 5.234]
print(numbers)
numbers[0] = 99
numbers[-1] = "Winter"
print(numbers)
```

```python
### 4. Adding Elements to a List
print(fruits)
print(numbers)
fruits.append("watermelon")
numbers.append(66)
print(fruits)
print(numbers)
```

```python
### 5. Removing Elements from a List
fruits = ["apple", "banana", "cherry" , 3.14]
numbers = [1, 2, 3, 4, 5 , "Python" , 5.234]

print(fruits)
fruits.remove("apple")
print(fruits)

print(numbers)
numbers.pop(2) #index
print(numbers)
```

```python
print(numbers)
numbers.pop(4)
print(numbers)
```

```python
### 6. Slicing a List
print(fruits)
some_fruits = fruits[1:3]
print(some_fruits)

print(numbers)
first_three_numbers = numbers[2:4]
print(first_three_numbers)
```

```python
### 6. Slicing a List
print(fruits)
some_fruits = fruits[-3:-1]
print(some_fruits)

print(numbers)
first_three_numbers = numbers[-4:-1]
print(first_three_numbers)
```

```python
### 6. Slicing a List
print(fruits)
some_fruits = fruits[-1:-3]
print(some_fruits)
some_fruits = fruits[-3:-1]
print(some_fruits)

print(numbers)
first_three_numbers = numbers[-4:-1]
print(first_three_numbers)
```

```python
### 7. Checking Membership in a List
print(fruits)
print(numbers)

is_apple_in_list = "apple" in fruits
print(is_apple_in_list)

is_seven_in_list = 7 in numbers
print(is_seven_in_list)

is_seven_in_list = 7 not in numbers
print(is_seven_in_list)
```

```python
### 8. List Length
print(fruits)
print(numbers)
num_fruits = len(fruits)
num_numbers = len(numbers)
print(num_fruits)
print(num_numbers)
```

```python
### 9. Concatenating Lists

print(fruits)
print(numbers)
colors = ['red', 'blue', 'green']
all_items = fruits + numbers + colors
print(all_items)
```

```python
### 11. Sorting a List
numbers = [10, 2, 4, 5, 6]
fruits= ['apple', "orange", 'blueberry', 'cherry' ]
print(numbers)
numbers.sort()
print(numbers)

print(fruits)
sorted_fruits = sorted(fruits)

print(fruits)
print(sorted_fruits)
```

```python
#13 Nested List
fruits= ['apple', "orange", 'blueberry', 'cherry' ]

# A bookshelf with shelves (nested lists)
bookshelf = [
    ["Book A", "Book B", "Book C"],   # Shelf 1,
    ["Book D", "Book E"],            # Shelf 2
    ["Book F", "Book G", "Book H"]   # Shelf 3
]

# Accessing items
print(bookshelf[0])      # First shelf -> ["Book A", "Book B", "Book C"]
print(bookshelf[0][1])   # Second book on the first shelf -> "Book B"

# # Adding a book to the second shelf
bookshelf[1].append("Book X")
print(bookshelf[1])  # ["Book D", "Book E", "Book X"]
```

```python
#14 Nested List - Access

bookshelf = [
    ["Book A", "Book B", "Book C"],   # Shelf 1
    ["Book D", "Book E"],            # Shelf 2
    ["Book F", "Book G", "Book H"]   # Shelf 3
]

# Loop through shelves
for shelf in bookshelf:
    print("Shelf:", shelf)  # Each shelf is a list
    # Loop through books in that shelf
    for book in shelf:
        print("   Book:", book)
```

```python
### 15. List Comprehension

squares = [x**2 for x in range(10)]
print(squares)


y=[]
for x in range(10):
  y.append(x**2)

print(y)
```

```python
### 16. Filtering a List with List Comprehension
numbers = [10, 2, 4, 5, 6]
even_numbers = [x for x in numbers if x % 2 == 0]
print(even_numbers)
```

```python
### 17. List Comprehension with Conditional Logic
numbers = [10, 2, 4, 5, 6]
new_list = [x if x % 2 == 0  else -1 for x in numbers]
print(new_list)
```

```python
### 18. Copying a List
print(fruits)
copied_fruits = fruits[:]
print(copied_fruits)
```

```python
### 19. Multidimensional Lists

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
```

```python
### 20. Flattening a Multidimensional List

flat_list = [item for sublist in matrix for item in sublist]
print(flat_list)
```
