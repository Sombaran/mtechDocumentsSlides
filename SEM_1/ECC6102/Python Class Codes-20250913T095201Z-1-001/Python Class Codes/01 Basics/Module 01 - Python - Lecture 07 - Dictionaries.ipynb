```python

# You can find a key by its value, but it's not as direct as a standard key lookup.
# Since multiple keys can have the same value, you need to iterate through the dictionary to find a match.

student_scores = {'Niraj': 95, 'Suraj': 88, 'Ramesh': 95, 'Clara': 76}
target_score = 95
found_key = None # Initialize a variable to store the key

print(student_scores['Clara'])
# Use .items() to get both the key and value for each pair
for key, value in student_scores.items():
  if value == target_score:
    found_key = key
    break # Stop the loop once the first match is found

print(f"The score {target_score} belongs to: {found_key}")
```

```python
original_dict  = {'Niraj': 94, 'Suraj': 88, 'Ramesh': 95, 'Clara': 76}

# Swapping keys and values using dictionary comprehension
swapped_dict = {value: key for key, value in original_dict.items()}

print(f"Original dictionary: {original_dict}")
print(f"Swapped dictionary: {swapped_dict}")
```

```python

# You can find a key by its value, but it's not as direct as a standard key lookup.
# Since multiple keys can have the same value, you need to iterate through the dictionary to find a match.

student_scores = {'Niraj': 95, 'Suraj': 88, 'Ramesh': 95, 'Clara': 99}
target_score = 95
found_key = None # Initialize a variable to store the key
all_keys = []

print(student_scores['Clara'])

# Use .items() to get both the key and value for each pair
for key, value in student_scores.items():
  if value == target_score:
    found_key = key
    all_keys.append(key)

print("Hello World")
# print("The score {target_score} belongs to: {found_key}")
print(f"The score {target_score} belongs to: {found_key}")
print(f"All keys having score as {target_score} are {all_keys}")
```

```python
from collections import defaultdict

d = defaultdict(int)  # Default value of int() is 0
print(d)

d['apple'] += 1
print(d['apple'])  # Output: 1
print(d['banana'])  # Output: 0 (default value assigned)
print(d)
```

```python
### 12. Using `collections.defaultdict` for Automatic Dictionary Values

from collections import defaultdict
fruits = ["apple", "banana", "apple", "orange","Pear", "Grapes", "banana", "apple", "Pear", "Grapes"]

fruit_count = defaultdict(int)
print(fruit_count)
for x in fruits:
    fruit_count[x] += 1

print(fruit_count)
```
