```python
### dictionary functions and methods in Python,

### 1. `clear()`: Removing All Items from a Dictionary

student = {"name": "Alice", "age": 21}

print(f"Before clear() {student}")
student.clear()
print(f"After clear() {student}")
```

```python
### 2. `copy()`: Creating a Shallow Copy of a Dictionary
import copy
original = {"name": "Alice", "age": 21}
shallow_copied = original.copy()

print(original)
print(shallow_copied)
print('*'*30)

shallow_copied["name"] = "Nitin"
print(original)
print(shallow_copied)
print('*'*30)
original["age"] = 99
print(original)
print(shallow_copied)
```

```python
### 3. `fromkeys()`: Creating a New Dictionary with Specified Keys

keys = ["name", "age", "major"]
default_value = "5"
new_dict = dict.fromkeys(keys, default_value)
print(new_dict)
```

```python
### 8. `pop()`: Removing and Returning a Value by Key

student = {"name": "Alice", "age": 21}
print(student)
key_to_be_removed = "age"
age = student.pop(key_to_be_removed, "N/A")

print(key_to_be_removed)
print(age)
print(student)
```

```python
### 9. `popitem()`: Removing and Returning the Last Key-Value Pair
student = {"name": "Alice",  "hobby2":"hockey", "age": 21, "hobby":"cricket", }
print(len(student))
print(f"Before popitem() {student}")



last_item = student.popitem()
print(last_item)

print(f"After popitem() {student}")
```

```python
### 11. `update()`: Updating Dictionary with Another Dictionary or Iterable
student = {"name": "Alice",  "hobby2":"hockey", "age": 21, "hobby":"cricket", }

student["fav_food"] = "Laddu"
print(f"{student}")
print("*"*30)
student["age"] = 99
print(student)

new_values = {"age": 45, "major": "CS"}
student.update(new_values)
print(student)
```

```python
### 17. `OrderedDict` from `collections`: Preserving Insertion Order

from collections import OrderedDict

ordered_dict = OrderedDict()

print(ordered_dict, type(ordered_dict))

ordered_dict["joker"] = 777
ordered_dict["value"] = 15
ordered_dict["score"] = 96
ordered_dict["rainbow"] = 20
ordered_dict["last"] = -95

print(ordered_dict, type(ordered_dict))
```

```python
from collections import OrderedDict

# Creating an OrderedDict
od = OrderedDict()
od["a"] = 1
od["b"] = 2
od["c"] = 3

print(od)  # OrderedDict([('a', 1), ('b', 2), ('c', 3)])

# Moving an item to the end
od.move_to_end("a")
print(od)  # OrderedDict([('b', 2), ('c', 3), ('a', 1)])

# Removing the last item
od.popitem()
print(od)  # OrderedDict([('b', 2), ('c', 3)])
```
