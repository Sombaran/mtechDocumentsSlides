```python
##### strings in Python

### 1. Simple String Assignment

greeting = "Hello"
name = "Sachin"
print(greeting+name)

print(greeting+" "+name)
```

```python
### 2. String Concatenation

full_greeting = greeting + ", " + name + "!"
print(full_greeting)
```

```python
### 3. Accessing Characters in a String
name = "Rahul Dravid"
first_letter = name[2]
last_letter = name[-1]


print(first_letter)
print(last_letter)

print(name[-5])
print(name[8])


print(name[3])
print(name[-9])
```

```python
### 4. String Length

length_of_name = len(name)
print(length_of_name)
```

```python
### 5. String Slicing
name = "Rahul Dravid"
X = name[1:7] #7-1 = 6 characters
print(X)

X = name[2:9] #9-2 = 7 characters
print(X)

X = name[10:12] #12 - 10 = 2 characters
print(X)
```

```python
### 5. String Slicing
name = "Rahul Dravid"
first_three_letters = name[:3]
print(first_three_letters)

last_two_letters = name[2:-7]
print(last_two_letters)

last_two_letters = name[-7:2]
print(last_two_letters)


print(name[:]) #Entire string
print(name)
```

```python
### 6. Changing Case

uppercase_name = name.upper()
lowercase_greeting = greeting.lower()
print(uppercase_name)
print(lowercase_greeting)
```

```python
### 7. Finding Substrings

print(name)
print(name[::-1])
position_of_letter = name.find("r")
print(position_of_letter)
print(greeting)
is_present = "Hello" in greeting
print(is_present)
```

```python
### 8. Replacing Substrings
print(greeting)
new_greeting = greeting.replace("Hello", "Hi")
print(new_greeting)
```

```python
### 9. Splitting Strings

sentence = "This#is#a#test"
words = sentence.split('#')
print(words)
```

```python
### 9. Splitting Strings

sentence = "This,is,a,test"
words = sentence.split(',')
print(words)
```

```python
### 10. Joining Strings

new_sentence = " ".join(words)
print(new_sentence)
```

```python
### 11. Stripping Whitespace
user_input = "  Hell        o              "
cleaned_input = user_input.strip()
print(cleaned_input)
```

```python
### 12. Checking String Content
name="Alice"
is_digit = "123".isdigit()
is_alpha = name.isalpha()
print(is_digit)
print(is_alpha)
```

```python
### 13. String Formatting with `format()`

formatted_string = "Hello, {}. You are {} years old.".format(name, 26)
print(formatted_string)
```

```python
### 13. String Formatting with `format()`

formatted_string = "Hello, {}. You are {} years old.".format(61, "Sunil")
print(formatted_string)
```

```python
### 14. String Interpolation with f-strings

age = 42
interpolated_string = f"Hello, {name}. You are {age} years old. {age} is just a number"
print(interpolated_string)
```

```python
### 15. Multi-line Strings

multi_line_string = """This is
a multi-line
string."""

print(multi_line_string)
```

```python
### 16. Escaping Special Characters

escaped_string = "He said, \"Python is awesome!\""
print(escaped_string)
```

```python
### 17. Raw Strings

file_path = r"c:\Desktop\Anam\new"
print(file_path)
```

```python
### 18. Reversing a String
name="Rahul Dravid"
reversed_name = name[::-1]
reversed_name = name[0:11:-1]
print(reversed_name)
```

```python
### 19. Counting Substrings
name="Alice Mathew"
count_of_e = name.find('e')
print(count_of_e)

name="Alice Mathew"
count_of_e = name.count('e')
print(count_of_e)
```

```python
### 20. Advanced String Formatting with Specifiers

pi = 3.1415
print("Pi to three decimal places: {pi:.3f}")
print(f"Pi to three decimal places: {pi:.3f}")
p = 98769

print(f"Pi to three decimal places:{p:13d}")
```
