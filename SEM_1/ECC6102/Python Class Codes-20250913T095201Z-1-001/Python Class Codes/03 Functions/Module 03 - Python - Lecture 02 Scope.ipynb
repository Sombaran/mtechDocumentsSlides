```python
#1. **Simple Global Variable**

x = 10  # Global variable

def print_global():
    global x
    print(x)

print("Global x = ", x)
print_global()
```

```python
#2. **Modifying a Global Variable Inside a Function**

x = 10  # Global variable

def modify_global():
    global x
    x = 20
    print(x)

print("Before" , x)
modify_global()
print("After" , x)
```

```python
# 3. **Using a Global Variable in Multiple Functions**

x = 5  # Global variable

def increase():
    global x
    x += 4

def display():
    global x
    print(x)

increase()
display()
```

```python
#4. **Global Variable with Conditional Logic**

x = 10  # Global variable

def check_value():
    global x
    if x > 5:
        x += 10
    else:
        x -= 5

check_value()
print(x)
```

```python
#5. **Global Variable in a Loop**

x = 0  # Global variable

def increment():
    global x
    for y in range(5):
        x += 1

increment()
print(x)
```

```python
#6. **Global Variable in a Nested Function**

x = 50  # Global variable

def outer_function():
    print("Value of X before calling inner_function()", x)
    print(x)
    y = 10
    def inner_function():
      global x
      print("Value of X inside inner_function()", x)
      x *= 2
      print("Value of X inside inner_function()", x)
      print("Value of y inside inner_function()", y)
    inner_function()

print("Value of X before calling outer_fun()", x)
outer_function()
# inner_function()
print("Value of X after calling outer_fun()", x)
```

Local Variable Examples

```python
#1. **Simple Local Variable**

def local_example(): # Removed extra indent
    x = 10  # Local variable
    print(x)

x=5
local_example()
print(x)
```

```python
#2. **Local Variable with Function Parameter**

def add(a, b):
    result = a + b  # Local variable
    return result

print(add(5, 3))
# print(result)
```

```python
#3. **Local Variable with Conditional Logic**

def check_number(n): # Removed extra indent
    if n > 0:
        result = "Positive"  # Local variable
    else:
        result = "Non-positive"
    return result

print(check_number(10))
print(check_number(-5))
```

```python
#4. **Local Variable in a Loop**

def count_up(n):
    # global count
    count = 0  # Local variable
    while count < n:
        print(count)
        count += 1

count = 10
count_up(5)
```

```python
#5. **Local Variable in Nested Functions**

def outer_function(): # Removed extra indent
    def inner_function():
        x = 20  # Local to inner_function
        print(x)
    inner_function()
    # print(x)  # This would raise an error

outer_function()
```

```python
#6. **Local Variables with Multiple Functions**

def set_value():
    value = 42  # Local variable
    return value

def print_value():
    print(set_value())

print_value()
```

```python
#7. **Local Variable and Exception Handling**

def divide(a, b): # Removed extra indent
    try:
        result = a / b  # Local variable
    except ZeroDivisionError:
        result = "Cannot divide by zero"
    return result

print(divide(10, 2))
print(divide(10, 0))
```

```python
def divide_numbers(num1, num2):
    try:
        result = num1 / num2
        print(f"Result: {result}")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
    except TypeError:
        print("Error: Invalid input type. Please enter numbers.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Test cases
divide_numbers(10, 2)      # Valid case
divide_numbers(10, 0)      # ZeroDivisionError
divide_numbers(10, 'a')    # TypeError
divide_numbers(10, None)   # TypeError
```
