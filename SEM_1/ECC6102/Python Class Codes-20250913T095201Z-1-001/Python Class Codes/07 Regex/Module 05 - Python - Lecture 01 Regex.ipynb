```python
### 4. **Matching Whitespaces**
# \s matches any whitespace character, including:

# space ( )

# tab (\t)

# newline (\n)

# carriage return (\r)

# form feed (\f)

# vertical tab (\v)

pattern = r'\s+'
text = 'Hello                       World'
match = re.split(pattern, text)
print(match)  # Output: ['Hello', 'World']

match = re.findall(pattern, text)
print(match)  # Output: ['Hello', 'World']
```

```python
#Capturing Group
import re

text = "Full Name: Ravi Kumar 9678336342       "
pattern = r'Full Name: ([A-Z][a-z]+) ([A-Z][a-z]+) (\d+) (\s+)'

match = re.search(pattern, text)
if match:
    print("First Name:", match.group(1))
    print("Last Name:", match.group(2))
    print("Phone :", match.group(3))
    print("Extra Space:", match.group(4))
```

```python
import re

def validate_name(name):
    # Define the regex pattern for typical Indian names
    pattern = r'^[A-Z][a-z]{2,}$'

    # Match the name with the regex
    if re.match(pattern, name):
        return True
    else:
        return False

# Test cases
names = [
    "Amit",
    "Suresh",
    "Priya",
    "raj",        # Invalid: starts with lowercase
    "Al",         # Invalid: less than 3 characters after first letter
    "Ram123"      # Invalid: contains digits
]

# Testing the function
for name in names:
    print(f"'{name}' is valid: {validate_name(name)}")
```

```python
import re

def validate_website(url):
    # Define the regex pattern for matching Indian website URLs
    pattern = r'^www\.[a-zA-Z0-9\-]+\.(com|in|co\.in)$'

    # Match the URL with the regex
    if re.match(pattern, url):
        return True
    else:
        return False

# Test cases
urls = [
    "www.irctc.co.in",
    "www.example.com",
    "www.mysite.in",
    "www.testsite.org",  # Invalid as it's not a valid Indian TLD
    "www.some-website.co.in",
    "www.example123.in"
]

# Testing the function
for url in urls:
    print(f"'{url}' is valid: {validate_website(url)}")
```

```python
import re

def validate_name(name):
    # Define the regex pattern for title and surname
    pattern = r'^(Mr|Mrs|Ms)\.?\s[A-Z][a-z]+$'

    # Match the input name with the regex
    if re.match(pattern, name):
        return True
    else:
        return False

# Test cases
names = [
    "Mr. Sharma",
    "Mrs. Rani",
    "Ms. Gupta",
    "Ms Sharma",  # Valid without period
    "Dr. Smith"   # Invalid, as "Dr" is not in the pattern
]

# Testing the function
for name in names:
    print(f"'{name}' is valid: {validate_name(name)}")
```

```python
import re

def validate_aadhaar(aadhaar):
    # Define the regex pattern for Aadhaar numbers
    pattern = r'^\d{4} ?\d{4} ?\d{4}$'

    # Match the Aadhaar number with the regex
    if re.match(pattern, aadhaar):
        return True
    else:
        return False

# Test cases
aadhaars = [
    "123456789012",        # Valid: no spaces
    "1234 5678 9012",      # Valid: spaces between groups
    "123 456789012",       # Invalid: first group has 3 digits
    "12345678 9012",       # Invalid: second group has 8 digits
    "1234 5678 90123"      # Invalid: 13 digits
]

# Testing the function
for aadhaar in aadhaars:
    print(f"'{aadhaar}' is valid?: {validate_aadhaar(aadhaar)}")
```

```python
# To validate a PAN (Permanent Account Number) card using regular expressions in Python, the format for a PAN card is usually as follows:
# It consists of 10 characters.
# The first five characters are uppercase letters (A-Z).
# The next four characters are digits (0-9).
# The last character is an uppercase letter (A-Z).

import re

def validate_pan(pan):
    # Define the regex pattern for PAN card
    pattern = r'^[A-Z]{5}[0-9]{4}[A-Z]{1}$'

    # Match the input with the regex
    if re.match(pattern, pan):
        return True
    else:
        return False

# Test cases
pan1 = "ABCDE1234F"
pan2 = "A1B2C3D4E5"

print(validate_pan(pan1))  # Should return True
print(validate_pan(pan2))  # Should return False
```

```python
import re

def validate_vehicle_registration(vehicle_number):
    # Define the regex pattern for vehicle registration numbers
    pattern = r'^[A-Z]{2}\d{2}[A-Z]{2}\d{4}$'

    # Match the vehicle number with the regex
    if re.match(pattern, vehicle_number):
        return True
    else:
        return False

# Test cases
vehicle_numbers = [
    "KA01AB1234",   # Valid
    "MH12XY5678",   # Valid
    "KA1234AB567",  # Invalid: incorrect format (digits in place of letters)
    "UP01AB123",    # Invalid: fewer digits in the registration number
    "ka01ab1234",   # Invalid: lowercase letters
    "KA01-AB1234"   # Invalid: incorrect format (hyphen included)
]

# Testing the function
for vehicle_number in vehicle_numbers:
    print(f"'{vehicle_number}' is valid?: {validate_vehicle_registration(vehicle_number)}")
```

```python
import re

# Summary of the GSTIN Format:
# The structure of the GSTIN is typically as follows:
# 2 digits for the state code.
# 5 uppercase letters for the business name.
# 4 digits for the unique business identifier.
# 1 uppercase letter for the nature of business.
# 1 character (either letter or digit) for further identification.
# The letter Z is a fixed character.
# 1 character (either letter or digit) to complete the identifier.

# Regular expression for matching GSTIN
pattern = r'\d{2}[A-Z]{5}\d{4}[A-Z]{1}[A-Z\d]{1}[Z]{1}[A-Z\d]{1}'

# Sample text containing GSTINs
text = """
Some GSTIN numbers for testing:
22ABCDE1234F1Z5
27XYZAB5678G2Z1
19ABCD5678H2Z9
"""

# Find all GSTINs in the given text
matches = re.findall(pattern, text)

# Print the matched GSTINs
print("Matched GSTINs:", matches)
```

```python
#Positive look ahead
#Let's say you want to match the word "apple", but only if it is followed by the word "pie".

import re

# Sample text
text = "I love apple pie. I like apple juice. I eat apple pie every day."

# Regular expression pattern to match "apple" only if it is followed by "pie"
pattern = r'apple(?= pie)'

# Finding all matches using re.findall
matches = re.findall(pattern, text)

# Printing the matches
print(matches)  # Output: ['apple', 'apple']
```

```python
#Example in Python that demonstrates the use of a negative lookbehind assertion in a regular expression.
# The task is to match the word "apple", but only if it is not preceded by the word "green".


import re

# Sample text
text = "I have a red apple. I have a green apple. I have a yellow apple."

# Regular expression pattern to match "apple" not preceded by "green"
pattern = r'(?<!green )apple'

# Finding all matches using re.findall
matches = re.findall(pattern, text)

# Printing the matches
print(matches)  # Output: ['apple', 'apple']
```
