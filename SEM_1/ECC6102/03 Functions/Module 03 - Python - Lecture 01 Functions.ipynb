```python
#2. **Function with Parameters**

def welcome_message(name):
   print(f"Welcome to the course, {name}!")

welcome_message("Preeti")
welcome_message("Aditya")
```

```python

#3. **Function with Return Value**

def sum_numbers(a, b, c):
   return a + b + c  #55

result = sum_numbers(12, 20, 23)
print(result)
print(sum_numbers(1,6,9))
```

```python
#4. **Function with Default Parameter**

def introduce(name="Anil"):
   print(f"My name is {name}.")

introduce("Deepa")
introduce()

introduce("Aditya")
```

```python
#5. **Function with Default Parameter**

def introduce(rainbow):
   print(type(rainbow))
   for name in rainbow:
     print(f"Hello, {name}. Welcome to Python")

names = ["Aditya", "Preeti", "Suresh"]
introduce(names)
```

```python
#4. **Function with List Passing**

def introduce(names):
   print(type(names))
   names[1] = "Sachin"
   for name in names:
     print(f"Hello, {name} . Welcome to Python")

names = ["Aditya", "Preeti", "Suresh"]
introduce(names)
print(names)
```

```python
#5. **Function with Keyword Arguments**

def display_info(name, roll, age):
   print(f"{name} is {age} years old and her roll number is {roll}")

display_info("Sita", 26, 99)
display_info("26", "Sita", "99")
display_info(age=30, name="Sita", roll="99")
```

```python
#6. **Variable Length Arguments (`*args`)**

def list_friends(*friends):
   print("My friends are:", friends)
   print(type(friends))
   print(len(friends))
   for i in friends:
    print(f"Hi, {i}")

list_friends("Arjun", "Vijay")
list_friends("Arjun", "Vijay", "Kavita", "Sachin", "Rahul", "Sourav")
```

```python
#7. **Function with Multiple Return Values**

def student_marks(maths, science):
   total = maths + science
   average = total / 2
   return total, average

total_marks, avg_marks = student_marks(85, 90)
print(f"Total: {total_marks}, Average: {avg_marks}")
```

```python

#8. **Anonymous (Lambda) Function**

double = lambda x: x ** 2

# def double(x):
#   return x ** 2

print(double(25))  # Output: 30
```

```python

#9. **Function with `**kwargs` (Keyword Arguments)**

def display_student_info(**info):
   print(type(info))
   for key, value in info.items():
	   print(f"{key}: {value}")

display_student_info(name="Rohit", age=22, subject="Math")
```

```python
#10. **Recursive Function**

def factorial(n):
   if n == 1:
     return 1
   else:
	   return n * factorial(n - 1) # 5 * factorial(4)

result = factorial(5)
print(f"Factorial of 5 is {result}")
```
