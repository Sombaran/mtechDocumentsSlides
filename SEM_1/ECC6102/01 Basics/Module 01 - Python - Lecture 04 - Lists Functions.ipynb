```python
### 19. `zip()`: Pairing Elements from Two Lists

names = ["Alice", "Bob", "Charlie", "Apple"]  #len = 4
ages = [25, 30, 35, 56, 78]    #len = 5
paired_list = list(zip(names, ages))
print(paired_list)
```

```python
### 20. `map()`: Applying a Function to All Elements in a List

numbers = [1, 2, 3, 4, 5, 6]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)

squares = []
for x in numbers:
    print(x**2)
    squares.append(x*x)

print(squares)
```

```python
import timeit

# Define the function to be applied
def double(x):
    return x * 2

# Define the input data
numbers = list(range(1000))

# Measure time for map
map_time = timeit.timeit('list(map(double, numbers))', globals=globals(), number=100000)

# Measure time for for loop
loop_time = timeit.timeit('for x in numbers: double(x)', globals=globals(), number=100000)

print(f"Time using map: {map_time:.5f} seconds")
print(f"Time using for loop: {loop_time:.5f} seconds")
```

A shallow copy creates a new object, but it does not create new copies of nested objects. Instead, it copies references to the nested objects. This means that changes to mutable nested objects in the copy will affect the original list.



```python
#Shallow Copy
import copy

original_list = [
    [1, 2, 3],
    [4, 5, 6]
    ]

print(original_list[0])
print(original_list[0][0])

print(original_list[1])
print(original_list[1][2])

shallow_copied_list = copy.copy(original_list)  #Shallow Copy

# #to what we understand
# shallow_copied_list = [
#     [99, 2, 3],
#     [4, 5, 6]
#     ]


shallow_copied_list[0][0] = 99  # Modifying the first sublist

print(original_list)  # [[99, 2, 3], [4, 5, 6]] (Affected)
print(shallow_copied_list)  # [[99, 2, 3], [4, 5, 6]]



original_list[0][1] = 88  # Modifying the first sublist

print(original_list)  # [[99, 2, 3], [4, 5, 6]] (Affected)
print(shallow_copied_list)  # [[99, 2, 3], [4, 5, 6]]
```

**Deep Copy**
A deep copy creates a completely independent copy, including nested objects. This means that changes in the copied list do not affect the original.

```python
#Deep Copy
import copy

original_list = [
    [1, 2, 3],
    [4, 5, 6]
    ]

deep_copied_list = copy.deepcopy(original_list)

deep_copied_list[0][0] = 42  # Modifying the first sublist

print(original_list)  # [[99, 2, 3], [4, 5, 6]] (Unchanged)
print(deep_copied_list)  # [[42, 2, 3], [4, 5, 6]]


original_list[0][1] = 99  # Modifying the first sublist

print(original_list)  # [[99, 2, 3], [4, 5, 6]] (Unchanged)
print(deep_copied_list)  # [[42, 2, 3], [4, 5, 6]]

```
