```python
# 1. **`len(tuple)`**: Returns the length of the tuple.

my_tuple = (55, 31, 88,88)
print(len(my_tuple))  # Output: 4

print(type(my_tuple))
```

```python
# 2. **`max(tuple)`**: Returns the maximum value in the tuple.

my_tuple = (55, 31, 88)
print(max(my_tuple))  # Output: 88
```

```python
# 3. **`min(tuple)`**: Returns the minimum value in the tuple.

my_tuple = (55, 31, 88)
print(min(my_tuple))  # Output: 31
```

```python
# 4. **`sum(tuple)`**: Returns the sum of all elements in the tuple.

my_tuple = (55, 31, 88)
print(sum(my_tuple))  # Output: 174
```

```python
# 5. **`any(tuple)`**: Returns `True` if at least one element in the tuple is true.

my_tuple = (0, -9, 0)
print(any(my_tuple))  # Output: True
```

```python
# 6. **`all(tuple)`**: Returns `True` if all elements in the tuple are true.

my_tuple = (1, 31, 88)
print(all(my_tuple))  # Output: True
```

```python
# 7. **`in` (membership test)**: Checks if an item exists in the tuple.

my_tuple = (55, 31, 88)
print(31 in my_tuple)  # Output: True
print(88 not in my_tuple)  # Output: False
print(3.14 not in my_tuple)  # Output: True
```

```python
#8 Reversing a Tuple
t = (1, 2, 3, 4, 2.5)
print((1, 2, 3, 4, 2.5)[::-1])  # Output: (4, 3, 2, 1)
```

```python
#9 Converting a String to a Tuple

s = "Python is a Programming Language"
t = tuple(s)
print(t)

s = "Python is a Programming Language"
t = list(s)
print(t)
```

```python
#10 Iterating Over a Tuple

t = ('apple', 'banana', 'cherry')
for item in t:
    print(item)


t = ['apple', 'banana', 'cherry']
for item in t:
    print(item)
```

```python
# 15. **`sorted(tuple)`**: Returns a sorted list of the tuple's elements.

my_tuple = (88, 31, 55)

print((my_tuple))
print(sorted(my_tuple))  # Output: [31, 55, 88]
print((my_tuple))
```

```python
# 16. **`enumerate(tuple)`**: Returns an enumerate object of the tuple, which yields pairs containing a count and values.

my_tuple = (55, 31, 88,)


for x in (my_tuple):
    print(x)


for index, value in enumerate(my_tuple, start=999):
    print(index, value)
    # Output:
    # 0 55
    # 1 31
    # 2 88
```

```python
# 17. **`zip(*tuples)`**: Aggregates elements from multiple tuples into tuples.

tuple1 = (55, 31, 88, 98)
tuple2 = ('a', 'b', 'c', 'd', 'e')
zipped = tuple(zip(tuple1, tuple2))
print(zipped)  # Output: ((55, 'a'), (31, 'b'), (88, 'c'))


zipped = tuple(zip(tuple2, tuple1))
print(zipped)  # Output: (('a', 55), ('b', 31), ('c', 88), ('d', 98))
```

```python
# 17. **`zip(*tuples)`**: Aggregates elements from multiple tuples into tuples.

tuple1 = ('India', 'Aus', 'UK')
tuple2 = ('New Delhi', 'Canberra', 'London', 'd', 'e')
zipped = list(zip(tuple1, tuple2))
print(zipped)
```

```python
# 18. **`reversed(tuple)`**: Returns an iterator that yields elements in the reverse order.

my_tuple = (55, 31, 88)
print(tuple(reversed(my_tuple)))  # Output: (88, 31, 55)
```

```python
 # 20 Removing Duplicates from a Tuple

t = (1, 2, 2, 3, 4, 4, 5, 2, 4,5,1,1,2,3,3)
t_unique = tuple(set(t))
print(t_unique)  # Output: (1, 2, 3, 4, 5) (Order may vary)
```

```python
# 21 Filtering Elements in a Tuple

t = (1, 2, 3, 4, 5, 6)
even_t = tuple(filter(lambda x: x % 2 != 0, t))
print(even_t)  # Output: (2, 4, 6)

odd_list = []
for element in t:
  if element % 2 != 0:
    odd_list.append(element)
    print(element)

print(tuple(odd_list))
```
